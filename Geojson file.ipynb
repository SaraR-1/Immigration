```python
def merge_geojson(input_path, output_path, clean):
    all_geo_json_d = os.listdir(input_path)
    temp = []
    
    if clean == True:
        for file in all_geo_json_d:
            with open(input_path+file) as f:
                data = json.load(f)
                for d in data['features']:
                    if d["geometry"]["type"] != "Point":
                        temp.append(d)
                        try:
                            print(d["properties"]["name:en"])
                        except KeyError:
                            try:
                                print(d["properties"]["wikipedia"])
                            except KeyError:
                                print(d["properties"])
    else:
        for file in all_geo_json_d:
            with open(input_path+file) as f:
                data = json.load(f)
                for d in data['features']:
                    temp.append(d)
        
                  
    with open(output_path, 'w') as outfile:
        json.dump(temp, outfile)
```

```python
merge_geojson("geo_json_origin_area/", "prova.geojson", clean = True)
```

```python
with open("prova.geojson") as f:
    data = json.load(f)
    
it_lat = 42.50
it_long = 12.50
my_map = folium.Map(location=[it_lat, it_long], zoom_start=3, tiles="openstreetmap")

for d in data:
    folium.GeoJson(
                d,
                style_function=lambda feature: {
                'fillColor': "red",
                'color': 'black',
                'weight': 1,
                'dashArray': '5, 5'
            }
            ).add_to(my_map)

my_map.save("prova.html")
```
